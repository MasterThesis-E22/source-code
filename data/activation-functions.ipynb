# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

# Functions

```python
# Sigmoidal
def sig(x):
  return 1/(1+np.exp(-x))

# Sigmoidal derivative
def dsig(x):
  return sig(x) * (1- sig(x))

# Hyperbolic Tangent
def hyp(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

# Hyperbolic derivative
def dhyp(x):
  return 1 - hyp(x) * hyp(x)

# Rectified Linear Unit
def relu(x):
  temp = [max(0,value) for value in x]
  return np.array(temp, dtype=float)

# Derivative for RELU
def drelu(x):
  temp = [1 if value>0 else 0 for value in x]
  return np.array(temp, dtype=float)

# Leaky Rectified Linear Unit
def lrelu(x):
  alpha = 0.2
  temp = [max(alpha*value,value) for value in x]
  return np.array(temp, dtype=float)

# Derivative for RELU
def dlrelu(x):
  temp = [1 if value>0 else 0.01 for value in x]
  return np.array(temp, dtype=float)
```

# Plot Functions

```python
# Generating data to plot
x_data = np.linspace(-5,5,1000)

fig, axes = plt.subplots(2, 2, width_ratios=[2, 2], figsize=(5,4))
ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[1,0]
ax4 = axes[1,1]

ax1.plot(x_data, sig(x_data), x_data, dsig(x_data))
ax2.plot(x_data, hyp(x_data), x_data, dhyp(x_data))
ax3.plot(x_data, relu(x_data), x_data, drelu(x_data))
ax4.plot(x_data, lrelu(x_data), x_data, dlrelu(x_data))

ax1.set_title('Sigmoid')
ax1.set_xlim([-5, 5])
ax1.set_xticks([], minor=True)
ax1.set_ylabel('y')

ax2.set_title('TanH')
ax2.set_xlim([-5, 5])
ax2.set_xticks(np.linspace(-5,5,3))
ax2.set_ylabel('y')

ax3.set_title('ReLu')
ax3.set_xlim([-5, 5])
ax3.set_xticks(np.linspace(-5,5,3))
ax3.set_xlabel('x')
ax3.set_ylabel('y')

ax4.set_title('Leaky ReLu')
ax4.set_xlim([-5, 5])
ax4.set_ylim([-2, 5])
ax4.set_xticks(np.linspace(-5,5,3))
ax4.set_xlabel('x')
ax4.set_ylabel('y')

fig.legend(["\u03C3(x)", "\u03C3'(x)"], bbox_to_anchor=(1.05,0.5), loc='center', ncol=1, bbox_transform=fig.transFigure)
fig.tight_layout()
```
