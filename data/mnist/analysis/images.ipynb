# Imports

```python
import os.path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image as Image
import idx2numpy
import cv2 as cv

def resize_image(image, width, height):
    newSize = (width, height)
    img = image.resize(newSize)
    return img
```

# Configs

```python
train_data_root = 'C:\\Users\\MLR\\data'
test_data_root = 'C:\\Users\\MLR\\data'
#train_data_root = '/mnt/data/mlr_ahj_datasets/mnist/packed/train'
#test_data_root = '/mnt/data/mlr_ahj_datasets/mnist/packed/test'
```

# Load data

```python
train_data = idx2numpy.convert_from_file(os.path.join(train_data_root, 'train-images-idx3-ubyte'))
train_targets = idx2numpy.convert_from_file(os.path.join(train_data_root, 'train-labels-idx1-ubyte'))

test_data = idx2numpy.convert_from_file(os.path.join(test_data_root, 't10k-images-idx3-ubyte'))
test_targets = idx2numpy.convert_from_file(os.path.join(test_data_root, 't10k-labels-idx1-ubyte'))
```

# Image Analysis

## Visualize one of each class

```python
df = pd.DataFrame(train_targets)

indexes = []
for i in range(0, 10):
    indexes.append(df[df[0] == i].index[0])
indexes

f, axarr = plt.subplots(2,5)


curr_row = 0
curr_col = 0

for i in indexes:
    if curr_col == 5:
        curr_col = 0
        curr_row = 1
    axarr[curr_row, curr_col].imshow(train_data[i], cmap='gray')
    axarr[curr_row, curr_col].set_axis_off()
    curr_col += 1

f.tight_layout(h_pad=0, w_pad=1, rect=(0, 0, 0.7, 0.7))
f.savefig("images/mnist_one_of_each.png", bbox_inches='tight', pad_inches=0)
```

## Sizing

```python
plt.figure(figsize = (4,4))
plt.imshow(train_data[0], cmap='gray')
plt.show()

plt.figure(figsize = (4,4))
plt.imshow(resize_image( Image.fromarray(train_data[0]), 14, 14), cmap='gray' )
plt.show()

plt.figure(figsize = (4,4))
plt.imshow(resize_image( Image.fromarray(train_data[0]), 7, 7), cmap='gray' )
plt.show()
```
